# Which are the sentences where "love" was used?

```python
pattern = re.compile("[A-Z]{1}[^.]*love[^a-zA-z]+[^.]*.")
findings = re.findall(pattern, book)
print(findings)
len(findings)
```

# What are the most used words?

```python
def get_word_count():
    pattern = re.compile("[a-zA-Z]+")
    words = re.findall(pattern, book.lower())
    dict = {}
    for word in words:
        if word in dict:
            dict[word] = dict[word] + 1
        else:
            dict[word] = 1
    return dict


word_count = get_word_count()
word_count = [(values, key) for (key, values) in word_count.items()]
results = sorted(word_count, reverse=True)
results
```

## Extract the paragraphs where "love" was used

## Extract the chapter titles

```python
pattern = re.compile("Chapter [0-9]+\n\n([^\n]+)")
titles = re.findall(pattern, book)
print(titles)
```

## Function that finds the occurrence of any word

```python
def get(word):
    word_count = get_word_count()
    print(word_count)
    if word in word_count:
        return word_count[word]
    else:
        return 0
```
